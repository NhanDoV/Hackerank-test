```python
from math import *

def area_by_edge(a, b, c):
    p = (a+b+c)/2
    return sqrt(p*(p-a)*(p-b)*(p-c))
area_by_edge(4, 5, 3)
```

```python
def edges_by_triangle_coordinates(list_of_triangle_coordinates):
    x1, y1 = list_of_triangle_coordinates[0]
    x2, y2 = list_of_triangle_coordinates[1]
    x3, y3 = list_of_triangle_coordinates[2]
    def single_edge(x1, y1, x2, y2):
        return sqrt( (x1 - x2)**2 + (y1 - y2)**2 )    
    return single_edge(x1, y1, x2, y2), single_edge(x1, y1, x3, y3), single_edge(x3, y3, x2, y2)
edges_by_triangle_coordinates([(0, 0), (4, 0), (2, 3)])
```

```python
import matplotlib.pyplot as plt

triangle_coordinates = [(0, 0), (4, 0), (2, 3), (0, 0)]
a, b, c = edges_by_triangle_coordinates(triangle_coordinates[1:])
plt.figure(figsize=(5, 5))
plt.plot([d[0] for d in triangle_coordinates], [d[1] for d in triangle_coordinates])
plt.plot([2, 2], [3, 0], 'green', alpha=0.5)
for x, y in triangle_coordinates:
    plt.plot(x, y, 'o', color='red')
plt.title(f"Base: 2, area: {area_by_edge(a, b, c):.2f}")
plt.text(2.05, 0.05, 'base', color='red')
plt.text(2.05, 1.5, 'height', color='blue')
plt.show()
```
