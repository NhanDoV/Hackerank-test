## [Normal distribution 1](https://www.hackerrank.com/challenges/normal-distribution-1/problem?isFullScreen=true)
Let $X \sim \mathcal{N}(\mu, \sigma)$ where $\mu = 30$ and $\sigma = 4$. Find
- a. $P(X  < 40)$
- b. $P(X > 21)$
- c. $P(30 < X < 35)$

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

muy = 30
sig = 4
data = np.arange(10,60,0.01)
pdf = norm.pdf(data , loc = muy , scale = sig )
fig, ax = plt.subplots(1, 3, figsize=(15, 4.5))
for _ in range(3):
    ax[_].plot(data, pdf)
    ax[_].set_xlabel('X')
    ax[_].set_ylabel('Probability Density')
    ax[_].text(x=27,y=0.05, s="\u03bc = 30 \n \u03C3 = 4")
    if _ == 0:
        ax[_].fill_between(data[data < 40], norm.pdf(data[data < 40], loc = muy , scale = sig ), facecolor='cyan' )
        ax[_].text(x=40, y=0.02, s='Prob(X < 40)')
        ax[_].set_title(f"a. Prob(X < 40): {norm(loc = muy , scale = sig).cdf(40):.3f}")
    elif _ == 1:
        ax[_].fill_between(data[data > 21], norm.pdf(data[data > 21], loc = muy , scale = sig ), facecolor='violet' )
        ax[_].text(x=21, y=0.02, s='Prob(X > 21)')
        ax[_].set_title(f"b. Prob(X > 21): {(1-norm(loc = muy , scale = sig).cdf(21)):.3f}")
    else:
        ax[_].fill_between(data[(data > 30) & (data < 35)], norm.pdf(data[(data > 30) & (data < 35)], loc = muy , scale = sig ), facecolor='orange' )
        ax[_].text(x=34, y=0.08, s='Prob(30 < X < 35)')
        ax[_].set_title(f"c. Prob(30 < X < 35) = P(X < 35) - P(X < 30): {(norm(loc = muy , scale = sig).cdf(35) - norm(loc = muy , scale = sig).cdf(30)):.3f}")
```

# Normal distribution 2
In a certain plant, the time taken to assemble a car is a random variable having a `normal distribution` with a `mean` of $20$ hours and a `standard deviation` of $2$ hours. What is the probability that a car can be assembled at this plant in:
- Less than $19.5$ hours ?
- Between $20$ and $22$ hours?s?

```python
muy = 20
sig = 2
data = np.arange(0, 40,0.01)
pdf = norm.pdf(data , loc = muy , scale = sig )
fig, ax = plt.subplots(1, 2, figsize=(15, 4.5))
for _ in range(2):
    ax[_].plot(data, pdf)
    ax[_].set_xlabel('X')
    ax[_].set_ylabel('Probability Density')
    ax[_].text(x=19,y=0.05, s="\u03bc = 30 \n \u03C3 = 2")
    if _ == 0:
        ax[_].fill_between(data[data < 19.5], norm.pdf(data[data < 19.5], loc = muy , scale = sig ), facecolor='cyan' )
        ax[_].text(x=25, y=0.02, s='Prob(X < 19.5)')
        ax[_].set_title(f"a. Prob(X < 19.5): {norm(loc = muy , scale = sig).cdf(19.5):.3f}")
    else:
        ax[_].fill_between(data[(data > 20) & (data < 22)], norm.pdf(data[(data > 20) & (data < 22)], loc = muy , scale = sig ), facecolor='violet' )
        ax[_].text(x=24, y=0.08, s='Prob(20 < X < 22)')
        ax[_].set_title(f"b. Prob(20 < X < 22) = P(X < 22) - P(X < 20): {(norm(loc = muy , scale = sig).cdf(22) - norm(loc = muy , scale = sig).cdf(20)):.3f}")
```

## Normal distribution 3

a) scored higher than 80?

b) should pass the test (grades≥60)?

c) should fail the test (grade
s<60)?<60)?

```python
muy = 70
sig = 10
data = np.arange(40,100,0.01)
pdf = norm.pdf(data , loc = muy , scale = sig )
fig, ax = plt.subplots(1, 3, figsize=(15, 4.5))
for _ in range(3):
    ax[_].plot(data, pdf)
    ax[_].set_xlabel('X')
    ax[_].set_ylabel('Probability Density')
    ax[_].text(x=65,y=0.03, s="\u03bc = 70 \n \u03C3 = 10")
    if _ == 0:
        ax[_].fill_between(data[data > 80], norm.pdf(data[data > 80], loc = muy , scale = sig ), facecolor='cyan' )
        ax[_].text(x=40, y=0.01, s='Prob(X > 80)')
        ax[_].set_title(f"a. Prob(X > 80): {(1-norm(loc = muy , scale = sig).cdf(80)):.5f}")
    elif _ == 1:
        ax[_].fill_between(data[data > 60], norm.pdf(data[data > 60], loc = muy , scale = sig ), facecolor='violet' )
        ax[_].text(x=21, y=0.01, s='Prob(X > 60)')
        ax[_].set_title(f"b. Prob(X > 60): {(1-norm(loc = muy , scale = sig).cdf(60)):.5f}")
    else:
        ax[_].fill_between(data[data < 60], norm.pdf(data[data < 60], loc = muy , scale = sig ), facecolor='orange' )
        ax[_].text(x=34, y=0.01, s='Prob(X < 60)')
        ax[_].set_title(f"c. Prob(X < 60): {norm(loc = muy , scale = sig).cdf(60):.5f}")
```
