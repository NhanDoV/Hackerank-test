Mary has an $n \times m$ piece of paper that she wants to cut into $1 \times 1$ pieces according to the following rules:
- She can only cut one piece of paper at a time, meaning she cannot fold the paper or layer already-cut pieces on top of one another.
- Each cut is a straight line from one side of the paper to the other side of the paper. For example, the diagram below depicts the three possible ways to cut a $3\times 2$ piece of paper



Given $n$ and $m$, find and print the minimum number of cuts Mary must make to cut the paper into $n \dot m$ squares that are $1 \times 1$ unit in size.

##### Input Format
A single line of two space-separated integers denoting the respective values of $n$ and $m$.

##### Output Format
Print a long integer denoting the minimum number of cuts needed to cut the entire paper into $1 \times 1$ squares.

```python
import math
import matplotlib.pyplot as plt

def cut(n, m):
    # Write your code here
    return n*m - 1

def illustration(n, m):
    data = [(x, y) for x in range(n+1) for y in range(m+1)]
    return data
```

```python
check_list = [(1, 3), (2, 3), (3, 3), (4, 5), (8, 9), (12, 12)]
fig, ax = plt.subplots(2, 3, figsize = (20, 7))
ax = ax.ravel()
for idx, data in enumerate(check_list):
    n, m = data
    z = illustration(n, m)
    for l in range(m+1):
        ax[idx].plot([0, n], [l, l], color='green')
    for k in range(n+1):
        ax[idx].plot([k, k], [0, m], color='green')
    ax[idx].plot([t[0] for t in z], [t[1] for t in z], 'o', color='red')
    ax[idx].set_title(f"{cut(n, m)} is the minimum numbers of cut the square {n}x{m} into 1x1")
    ax[idx].set_axis_off()
plt.show()
```

Noting that while cutting these squares will following these rules:
- We can **only cut one piece of paper at a time**, meaning she cannot fold the paper or layer already-cut pieces on top of one another.
- **Each cut is a straight line from one side of the paper to the other side** of the paper.
