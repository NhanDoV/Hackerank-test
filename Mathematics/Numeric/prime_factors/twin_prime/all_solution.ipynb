```python
def is_prime(x):
    """
        We know that the unique even number is a prime is 2 so we will ignore all even.number which is greater than 2
        Loop only on the odd numbers from 3 to x to find the division-factors
        Time complexity: O(n)
    """
    prime_ls = []
    for k in range(3, x, 2):
        if x % k == 0:
            prime_ls.append(k)
    if ((len(prime_ls) > 0) or (x % 2 == 0) and x!= 2) or (x==1):
        print(f"{x} is not a prime since its divisions are {prime_ls}")
        return False
    else:
        print(f"{x} is a prime")
        return True

for num in check_list[:4]:
    is_prime(num)
```

### 1. Check if a twin

```python
def is_prime(x):
    """
        Loop only on the odd numbers from 3 to 10*log(x) to find the division-factors
        Time complexity: O(log(n))
    """
    prime_ls = []
    for k in range(3, 100*int(log(x))+1, 2):
        if (x % k == 0) and (k != x):
            prime_ls.append(k)
    if ((len(prime_ls) > 0) or (x % 2 == 0) and x!= 2) or (x==1):
        return False
    else:
        return True
    
def check_twins(x, y):
    if is_prime(y) and is_prime(x) and abs(x - y) == 2:
        print(f"{x} and {y} is a twins @-@")
    else:
        print(f"{x} and {y} is not a twins!!!!!")

check_list = [(3, 5), (17, 19), (91, 93), (10007, 10009), (1000000007, 1000000009)]
for nums in check_list:
    x, y = nums
    check_twins(x, y)
```

### 2. Check all prime in a range of numbers

```python
def is_prime(x):
    """
        Loop only on the odd numbers from 3 to 10*log(x) to find the division-factors
        Time complexity: O(log(n))
    """
    prime_ls = []
    for k in range(3, max(500, int(log(x))+1), 2):
        if (x % k == 0) and (k != x):
            prime_ls.append(k)
            break
    if ((len(prime_ls) > 0) or (x % 2 == 0) and x!= 2) or (x==1):
        return False
    else:
        return True

def list_primes(a, b):
    """
        List out all of primes in ranged of (a, b)
        If num is a primes, we will find the next elements as type of next_num = num + 2^n
    """    
    if a < 2:
        primes_ls = [2]
    else:
        primes_ls = []
            
    if a % 2 == 0:
        a = a + 1
    if b % 2 == 0:
        b = b - 1

    for num in range(a, b+1, 2):
        if is_prime(num):
            primes_ls.append(num)
            
    return len(primes_ls), primes_ls

list_primes(1, 10000001)
```

```python
def list_primes(a, b):
    """
        List out all of primes in ranged of (a, b)
        If num is a primes, we will find the next elements as type of next_num = num + 2^n
    """
    if a % 2 == 0:
        a = a + 1
    if b % 2 == 0:
        b = b - 1
    primes_ls = []

    start = a
    while start <= b:
        if is_prime(start):
            primes_ls.append(start)
            start += 2
        else:
            start += 4
    return len(primes_ls), primes_ls

list_primes(1000000, 10000000)
```
